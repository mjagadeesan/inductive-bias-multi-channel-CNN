```python
import torch
data_tr = torch.load("training-test-data/training_data.txt")
target_tr = torch.load("training-test-data/training_targets.txt")
data_test = torch.load("training-test-data/test_data.txt")
target_test = torch.load("training-test-data/test_targets.txt")

num_samples = 512 # number of training samples
num_samples_test = 100 # number of test samples
new_dim1 = 32 * 1 # first dimension
new_dim2 = 32 * 1 # second dimension
old_dim = 32 # CIFAR original dimension
```

```python

# find the minimum norm solution 

import cvxpy as cp
import math


# beta_cv_pred = np.zeros((new_dim1, new_dim2))

def solve_cvxpy(option):
    global beta_cv_pred_0;
    global beta_cv_pred_1;
    global beta_cv_pred_2;
    data_tr_numpy =data_tr.data.numpy()
    target_tr_numpy =target_tr.data.numpy()

    print(data_tr.shape)
    print(target_tr.shape)

    dft_matrix1 = np.fft.fft(np.eye(new_dim1)) / np.sqrt(new_dim1)
    dft_matrix2 = np.fft.fft(np.eye(new_dim2)) / np.sqrt(new_dim2)

    beta_cv_0 = cp.Variable((new_dim1, new_dim2))
    beta_cv_1 = cp.Variable((new_dim1, new_dim2))
    beta_cv_2 = cp.Variable((new_dim1, new_dim2))
    beta_cv = cp.Variable((new_dim1 * new_dim2, 3)) # Only going to used for ell2ell1 
    function = 0 

  # ell2-ell1 norm

    if option == "ell2ell2":
        function += 2 * cp.norm(beta_cv, 'fro')

    if option == "ell2ell1":
        for z in range(new_dim1 * new_dim2):
            function += 2 * cp.norm(beta_cv[z]) 

  # Sum of the ell1 norms
    if option == "ell1":
        function += 2 * cp.sum(cp.abs(dft_matrix1 @ beta_cv_0 @ dft_matrix2.T))
        function += 2 * cp.sum(cp.abs(dft_matrix1 @ beta_cv_1 @ dft_matrix2.T))
        function += 2 * cp.sum(cp.abs(dft_matrix1 @ beta_cv_2 @ dft_matrix2.T))
    objective = cp.Minimize(function)
  
  # Set up constraints
    constraints = [target_tr[i] * cp.sum(cp.multiply(beta_cv_0, data_tr_numpy[i][0]) + cp.multiply(beta_cv_1, data_tr_numpy[i][1]) + cp.multiply(beta_cv_2, data_tr_numpy[i][2])) >= 1 for i in range(num_samples)] # margin >=1 
    if option == "ell2ell1" or option == "ell2ell2":
        beta_cv_fft_0 = cp.vec(cp.abs(dft_matrix1 @ beta_cv_0 @ dft_matrix2.T))
        beta_cv_fft_1 = cp.vec(cp.abs(dft_matrix1 @ beta_cv_1 @ dft_matrix2.T))
        beta_cv_fft_2 = cp.vec(cp.abs(dft_matrix1 @ beta_cv_2 @ dft_matrix2.T))
        beta_cv_tr = beta_cv.T
        constraints = constraints + [beta_cv_fft_0 <= beta_cv_tr[0]]
        constraints = constraints + [beta_cv_fft_1 <= beta_cv_tr[1]]
        constraints = constraints + [beta_cv_fft_2 <= beta_cv_tr[2]]

    prob = cp.Problem(objective, constraints)
    print("set up problem")
    result = prob.solve(solver=cp.SCS)
    print("solved problem")
    print(result)
    beta_cv_pred0 = beta_cv_0.value
    beta_cv_pred1 = beta_cv_1.value
    beta_cv_pred2 = beta_cv_2.value
    betas = [beta_cv_pred0, beta_cv_pred1, beta_cv_pred2]
    return betas




# # beta_cv_pred_ell1 = solve_cvxpy("ell1")
# # data_np =data_tr.data.numpy()
# # preds = [np.tensordot(data_np[i], beta_cv_pred) for i in range(num_samples)]
# # margins = [target_tr[i] * preds[i] for i in range(num_samples)]
# # max_margin = max(margins).data.numpy()
# # print(max_margin)

# plt.imshow(beta_cv_pred_ell2, cmap='gray', interpolation='none')
# plt.show()

# beta_cv_pred_ell1 = solve_cvxpy("ell1")

# plt.imshow(beta_cv_pred_ell1 , cmap='gray', interpolation='none')
# plt.show()
```
